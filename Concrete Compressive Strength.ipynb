### Outlier Detection and Removal
1. Detection can be done using IQR, Box plot, Percentile method, z score method and each technic has its own method of 
    removing the outliers
2. We will use IQR method to find and use Capping method to treat them
3. In Capping methods we replace the outliers with median as the median is not affected by outliers.
4. There is other method called trimming in which we delete the outliers.

```python
#### outlier treatment of age as the column age as many outlier values
df = concrete_data
df.describe()
# Finding the IQR
percentile25 = df['Age'].quantile(0.25)
percentile75 = df['Age'].quantile(0.75)

print('percentile75',percentile75)
print('percentile25',percentile25)

iqr = percentile75 - percentile25
print('iqr',iqr)

upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr

print("Upper limit",upper_limit)
print("Lower limit",lower_limit)

#Finding Outliers
df[df['Age'] > upper_limit].shape
#Finding Outliers
df[df['Age'] < lower_limit].shape

# finding median of age
median = df.loc[df['Age'] < upper_limit, 'Age'].median()
print('median',median)

#Capping in this we replace the outliers with some value if true, if false we keep data as it is
new_df_cap = df.copy()

new_df_cap['Age'] = np.where(
    new_df_cap['Age'] > upper_limit,
    median,
    np.where(
        new_df_cap['Age'] < lower_limit,
        median,
        new_df_cap['Age']
    )
)

new_df_cap.describe()

sns.boxplot(data = new_df_cap, x =new_df_cap['Age'])
plt.title('Box plot after the outlier treatment')
plt.show()

# replacing new_df_cap in to original data frame ie concrete_data
concrete_data = new_df_cap
concrete_data.describe()
```

### Deciding the Type of problem
1. This problem is of the Regression type as the target variable is numeric(continuous) in nature.
2. The distribution of the data is not in linear nature and therefore we cannot use linear machine model which assumes the 
    data to be linearly distributed with respect to target variable.
3. The models which donot assume linear relation between independent and dependent variables like Decision Tree, Random Forest
    and support vector machine with kernal transformation may work well with this data set.
4. The models based on tree type structure like random forest and decision tree are robust to outliers. 
5. Decision Tree is more prone to overfitting and therefore Random forest will work good.

## Defining input and output for the model

```python
# defining input and output for the model
inputs = concrete_data.drop(['Concrete_compressive_strength'], axis = 1)
outputs = concrete_data['Concrete_compressive_strength']
print('the shape of the data is', concrete_data.shape)
print('the shape of inputs is', inputs.shape)
print('the shape of outputs is', outputs.shape)
```

## Dividing the data set into training and test sets

```python
# dividing the data set into trainig and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, outputs, test_size=0.30, random_state=1)

# checking the shape of training and test data set
print('The shape of X_train', X_train.shape)
print('The shape of X_test' , X_test.shape)
print('The shape of y_train', y_train.shape)
print('The shape of y_test' , y_test.shape)
```

### Applying standardization

```python
# applying standardization
from sklearn.preprocessing import StandardScaler
stn = StandardScaler()
X_train = stn.fit_transform(X_train)
X_test = stn.transform(X_test)
```

### Traing the model

```python
# we train the model by giving x train and y train.
# we train model using the training data set
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Ridge Regression
from sklearn.linear_model import Ridge
reg_ridge = Ridge()
reg_ridge.fit(X_train, y_train)

# Lasso Regression
from sklearn.linear_model import Lasso
reg_lasso = Lasso()
reg_lasso.fit(X_train, y_train)

# Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
reg_dt = DecisionTreeRegressor()
reg_dt.fit(X_train, y_train)

# Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
reg_rfg = RandomForestRegressor()
reg_rfg.fit(X_train, y_train)

# KNN Regressor
from sklearn.neighbors import KNeighborsRegressor
reg_knn = KNeighborsRegressor()
reg_knn.fit(X_train, y_train)

# SVM
from sklearn.svm import SVR
reg_svr = SVR()
reg_svr.fit(X_train, y_train)
```

### Predicting model output on test data

```python
# we use X_test ie test data set to do prediction after training the model on x train and y train data
# we store prediction in y_pred variable
y_pred_linear = lin_reg.predict(X_test)

# Predicting with Ridge Regression
y_pred_ridge = reg_ridge.predict(X_test)

# Predicting with Lasso Regression
y_pred_lasso = reg_lasso.predict(X_test)

# Predicting the Decision Tree Regressor
y_pred_dt = reg_dt.predict(X_test)

# Predicting with Random Forest Regressor
y_pred_rfg = reg_rfg.predict(X_test)

# Prediction using KNN Regressor
y_pred_knn = reg_knn.predict(X_test)

# Prediction using SVR
y_pred_svr = reg_svr.predict(X_test)
```

### Model Evaluation Mean squared error - Test Accuracy

```python
# for evaluation of the model we use y_test and y_pred

# linear Regression Evaluation
from sklearn.metrics import mean_squared_error
TS_MSE_linear = mean_squared_error(y_test, y_pred_linear)
print('The mean sqaured error for linear regression is:',TS_MSE_linear)

# Ridge Regression Evaluation
from sklearn.metrics import mean_squared_error
TS_MSE_ridge = mean_squared_error(y_test, y_pred_ridge)
print('The mean sqaured error for ridge regression is:',TS_MSE_ridge)

# Lasso Regression Evaluation
from sklearn.metrics import mean_squared_error
TS_MSE_lasso = mean_squared_error(y_test, y_pred_lasso)
print('The mean sqaured error for lasso regression is:',TS_MSE_lasso)

# Decision Tree Regression Evaluation
from sklearn.metrics import mean_squared_error
TS_MSE_dt = mean_squared_error(y_test, y_pred_dt)
print('The mean sqaured error for decision tree regression is:',TS_MSE_dt)

# Random Forest Regression Evaluation
from sklearn.metrics import mean_squared_error
TS_MSE_rfg = mean_squared_error(y_test, y_pred_rfg)
print('The mean sqaured error for random forest regression is:',TS_MSE_rfg)

# KNN Regression Evaluation
from sklearn.metrics import mean_squared_error
TS_MSE_knn = mean_squared_error(y_test, y_pred_knn)
print('The mean sqaured error for knn regression is:',TS_MSE_knn)

# SVM Regression Evaluation
from sklearn.metrics import mean_squared_error
TS_MSE_svr = mean_squared_error(y_test, y_pred_svr)
print('The mean sqaured error for svr regression is:',TS_MSE_svr)
```

### Multicolinearity

1. It is defined as the correlation between the independent variables.
2. Having the correlation between the independent variables is not good for the model as the model assumes the input variables 
    to be independent.
3. Having the correlation in input variables does not give the flexibility to modify any one input variable as they tend to 
    affect the other input variable due to dependency.
4. Having correaltion causes to loose the independency among the input variables and this will lead to redundant data 
    for the model to train and increases the computational time and cost.
5. The correlation among input variables decreases the importance of regression coefficient as the regression coefficient 
    describe the relation of input variables with the output variables.
6. Multicolinearity causes the curse of dimentionality. It increases the no.of dimention of data.
    
### Methods to Identify multicolinearity
1. Using the correlation heat map on the input variables.
2. Using the VIF variance inflation factor to determine the strength of correlation.
3. A value of VIF > 5 (Theoritically) or VIF > 10 (as per industry standards) is treated as the strong correlation and requires
    to be dealt it with either dropping them, standardizing the data, performing PCA.
